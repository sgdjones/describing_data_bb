```python
# our boilerplate

import pandas as pd
import numpy as np
```

## Initial standard calculations of the Brady's ages

```python
#First I created a data frame

bb_df = pd.DataFrame()

bb_df['name'] = ["Greg", "Marcia", "Peter", "Jan", "Bobby", "Cindy", "Oliver"]
bb_df['age'] = [14, 12, 11, 10, 8, 6, 8]
bb_df.set_index("name")

```

```python
bb_df.describe()
```

```python
print (bb_df.median())
print (bb_df.mode())
```

```python
print (bb_df.var())
```

My inital calculations by hand were not far off.  
 - Mean at 9.857
 - Median at 10
 - Mode at 8
 
However, on the second set of calulations (σ2, σ, and standard error), I had transposed the numbers and calculated using a mean of 9.875 (typical).
 

## Calculations with Cindy's Birthday & Oliver replaced by Jessica

The mode did not change when Cindy turned 7 nor did the median. However, when Oliver left all the calculations changed including the mode and median, the averger was changed significantly as was the std.

```python
bb_bday_df = bb_df.replace(6,7)
```

```python
print(bb_bday_df)
```

```python
bb_bday_df.describe()
```

```python
bb_bday_df.var()
```

```python
bb_jessica = bb_bday_df
bb_jessica.drop([6], axis=0)
bb_jessica.append({"name":"Jessica", "age":1}, ignore_index=True)
```

```python
bb_jessica.describe()
```

```python
#calculating the standard error after Cindy's birthday
np.std(bb_bday_df['age'])/np.sqrt(len(['age']))
```

```python
#calculating the standard error with the replacement of Oliver w/ Jessica
np.std(bb_jessica['age'])/np.sqrt(len(["age"]))
```
